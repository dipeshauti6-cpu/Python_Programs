```python
'''Question 14:Write a python program to calculate sum of all even numbers for a given range.Given range: 1 - 50'''
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print("Sum =", total)
```

```python
'''Question 16:Write a python program to print the colours of fruits from the given list using for loop.'''
fruits = ["Apple", "Banana", "Grapes"]
colors = ["Red", "Yellow", "Green"]
for i in range(len(fruits)):
    print(fruits[i], "is", colors[i])
```

```python
'''Question 17:Find the minimum number present in the given list of numbers using for loop.'''
l = [22,64,7,89,6]
minimum = l[0]
for i in l:
    if i < minimum:
        minimum = i

print("Minimum number =", minimum)
```

```python
'''Question 18:Write a program to print first 10 natural numbers using while loop.'''
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
'''Question 19:Write a program to print numbers from 10 to 1.'''
i = 10
while i >= 1:
    print(i)
    i -= 1
```

```python
'''Question 20:Write a program to calculate sum of first 20 natural numbers using while loop.'''
i = 1
total = 0
while i <= 20:
    total += i
    i += 1
print("Sum =", total)
```

```python
'''Question 21:Print the program for Square of 1 to 10 numbers.'''
i = 1
while i <= 10:
    print(i, "=", i * i)
    i += 1
```

```python
'''Question 22:Print fibonacci series using for loop'''
n = 8
a = 0
b = 1
print(a)
print(b)
for i in range (n+1):
    c = a+b
    print(c)
    a=b
    b=c    
```
