```python
'''29. Write a Python program to print the pattern:
* 
** 
*** 
**** 
*****'''
for i in range(1,6):
    for j in range(1,6):
        if j<=i:
            print("*",end= " ")
        else:
            print(" ",end= " ")
    print()
```

```python
'''30. Write a Python program to print the pattern:
***** 
**** 
*** 
** 
*'''
for i in range(1,6):
    for j in range(1,6):
        if j<=6-i:
            print("*",end = " ")
        else:
            print(" ",end= " ")
    print()
```

```python
'''31. Write a Python program to print the pattern:
1
12
123
1234
12345'''
for i in range(1,6):
    for j in range(1,6):
        if j<=i:
            print(j,end=" ")
        else:
            print(" ",end=" ")
    print()
        
```
