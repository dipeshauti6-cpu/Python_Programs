```python
'''Question 5:Write a program to calculate the electricity bill.
First 100 units = No charge
Next 100 units = Rs 5 per unit
Above 200 units = Rs 10 per unit'''
unit = int(input("enter the unit"))
if unit<=100:
    print("Bill=",0)
elif unit<=200:
    print("Bill=",(unit-100)*5)
else:
    print("Bill=",(100*5)+(unit-200)*10)

```

```python
'''Question 11:Write a Python program to calculate profit or loss based onselling cost and actual cost.'''
cost_price = float(input("Enter Cost Price"))
selling_price = float(input("Enter Selling Price"))
if selling_price > cost_price:
    profit = selling_price - cost_price
    print("Profit =", profit)
else:
    loss = cost_price - selling_price
    print("Loss =", loss)
```

```python
'''Question 12:Ask user to enter salary and years of experience.
Offer 10% bonus if experience is greater than 5 years.
Print bonus amount and new salary.'''
salary = float(input("Enter Salary: "))
experience = int(input("Enter Years of Experience: "))
if experience > 5:
    bonus = salary * 10 / 100
    new_salary = salary + bonus
    print("Bonus Amount =", bonus)
    print("New Salary =", new_salary)
else:
    print("No Bonus")
    print("Salary =", salary)
```

```python
'''Question 13:Take prices of 4 products from user.Offer 10% discount if bill is greater than 5000,otherwise offer 5% discount.
Print discount amount and revised bill.'''
p1 = int(input("Enter Price of Product 1"))
p2 = int(input("Enter Price of Product 2"))
p3 = int(input("Enter Price of Product 3"))
p4 = int(input("Enter Price of Product 4"))
total = p1 + p2 + p3 + p4
print("Total Bill =", total)
if total > 5000:
    discount = total * 10 / 100
    print("Discount =", discount)
else:
    discount = total * 5 / 100
    print("Discount =", discount)

final_bill = total - discount
print("Final Bill =", final_bill)

```

```python
'''Question 17:Ask user to enter marks and print the corresponding grade.
Below 25 = F
25 to 45 = E
45 to 50 = D
50 to 60 = C
60 to 80 = B
Above 80 = A'''
marks = int(input("Enter Marks"))
if marks < 25:
    print("Grade F")
elif marks <= 45:
    print("Grade E")
elif marks <= 50:
    print("Grade D")
elif marks <= 60:
    print("Grade C")
elif marks <= 80:
    print("Grade B")
else:
    print("Grade A")
```

```python
'''Question 18:
A student will not be allowed to sit in exam
if attendance is less than 75%

Take input:
1. Number of classes held
2. Number of classes attended
Print percentage of attendance and eligibility.'''

held = int(input("Enter Number of Classes Held"))
attended = int(input("Enter Number of Classes Attended"))
percentage = (attended / held) * 100
print("Attendance Percentage =", percentage)
if percentage >= 75:
    print("Allowed to Sit in Exam")
else:
    print("Not Allowed to Sit in Exam")
```
