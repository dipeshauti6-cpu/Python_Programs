```python
marks = np.random.randint(0, 101, 50)
print("Above 75 =", np.sum(marks > 75))
print("Below 35 =", np.sum(marks < 35))
print("Between 35 and 75 =", np.sum((marks >= 35) & (marks <= 75)))
```

```python
marks = np.array([95, 80, 65, 50, 30])
for i in marks:
    if i >= 75:
        print(i, "A")
    elif i >= 60:
        print(i, "B")
    elif i >= 35:
        print(i, "C")
    else:
        print(i, "Fail")
```

```python
marks = np.array([80, 75, 90, 85, 88])
total = np.sum(marks)
average = np.mean(marks)
percentage = average
if percentage >= 75:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 35:
    grade = "C"
else:
    grade = "Fail"
print("Marks =", marks)
print("Total =", total)
print("Average =", average)
print("Percentage =", percentage)
print("Grade =", grade)
```
