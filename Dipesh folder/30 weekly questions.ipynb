```python
f = open("file1.txt", "w")
f.write("Name: Dipesh, Roll No: 101, Marks: 85\n")
f.write("Name: Amit, Roll No: 102, Marks: 60\n")
f.write("Name: Rahul, Roll No: 103, Marks: 80\n")
f.write("Name: Priya, Roll No: 104, Marks: 56\n")
f.write("Name: Neha, Roll No: 105, Marks: 88\n")
f.close()
f = open("file1.txt", "r")
lines = f.readlines()
f.close()
students = []
for line in lines:
    data = line.strip().split(", ")
    marks = int(data[2].replace("Marks: ", ""))
    students.append([marks, line])
students.sort()
print("Student Records Sorted by Marks:\n")
for student in students:
    print(student[1], end="")
```

```python
f = open("file1.txt", "r")
lines = f.readlines()
f.close()
above90 = 0
below35 = 0
for line in lines:
    marks = int(line.split(", ")[2].replace("Marks: ", ""))
    if marks > 90:
        above90 += 1
    if marks < 35:
        below35 += 1
print("Students scoring more than 90=", above90)
print("Students scoring less than 30=", below35)
```

```python
while True:
    print("1.Add\n2.View\n3.Search\n4.Update\n5.Delete\n6.Exit")
    ch = int(input("Enter your choice: "))
    if ch == 1:
        name = input("Enter Name: ")
        roll = input("Enter Roll No: ")
        marks = input("Enter Marks: ")
        f = open("file1.txt", "a")
        f.write("Name: " + name + ", Roll No: " + roll + ", Marks: " + marks + "\n")
        f.close()
    elif ch == 2:
        f = open("file1.txt", "r")
        print(f.read())
        f.close()
    elif ch == 3:
        roll = input("Enter Roll No: ")

        f = open("file1.txt", "r")
        for line in f:
            if "Roll No: " + roll in line:
                print(line)
        f.close()
    elif ch == 4:
        roll = input("Enter Roll No: ")
        old = input("Enter Old Marks: ")
        new = input("Enter New Marks: ")
        f = open("file1.txt", "r")
        data = f.read()
        f.close()
        data = data.replace("Roll No: " + roll + ", Marks: " + old,
                            "Roll No: " + roll + ", Marks: " + new)
        f = open("file1.txt", "w")
        f.write(data)
        f.close()
    elif ch == 5:
        roll = input("Enter Roll No: ")
        f = open("file1.txt", "r")
        lines = f.readlines()
        f.close()
        f = open("file1.txt", "w")
        for line in lines:
            if "Roll No: " + roll not in line:
                f.write(line)
        f.close()
    elif ch == 6:
        print("Program Ended")
        break
    else:
        print("Invalid Choice")
```

```python
import math
num = int(input("Enter a number: "))
power = int(input("Enter power: "))
f = open("file1.txt", "w")
f.write("Square Root = " + str(math.sqrt(num)) + "\n")
f.write("Factorial = " + str(math.factorial(num)) + "\n")
f.write("Power = " + str(math.pow(num, power)) + "\n")
f.close()
f = open("file1.txt", "r")
print(f.read())
f.close()
```
