```python
'''Question 1: Write a program to calculate the perimeter of a square.'''
side = float(input("Enter side of square: "))
perimeter = 4 * side
print("Perimeter =", perimeter)
```

```python
'''Question 2: Write a program to convert kilometers into meters.'''
km = float(input("Enter distance in kilometers: "))
meters = km * 1000
print("Distance in meters =", meters)
```

```python
'''Question 3: Write a program to convert minutes into hours.'''
minutes = int(input("Enter minutes: "))
hours = minutes / 60
print("Hours =", hours)
```

```python
'''Question 4: Write a program to calculate simple interest.'''
p = float(input("Enter principal amount: "))
r = float(input("Enter rate of interest: "))
t = float(input("Enter time: "))
si = (p * r * t) / 100
print("Simple Interest =", si)
```

```python
'''Question 5: Write a program to calculate the area of a triangle.'''
base = float(input("Enter base: "))
height = float(input("Enter height: "))
area = 0.5 * base * height
print("Area of Triangle =", area)
```

```python
'''Question 6: Write a program to find the remainder when one number is divided by another.'''
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
remainder = a % b
print("Remainder =", remainder)
```

```python
'''Question 7: Write a program to print numbers from 50 to 60.'''
for i in range(50, 61):
    print(i)
```

```python
'''Question 8: Write a program to print all numbers divisible by 3 from 1 to 30.'''
for i in range(1, 31):
    if i % 3 == 0:
        print(i)
```

```python
'''Question 9: Write a program to count the number of uppercase letters in a string'''
text = input("Enter a string: ")
count = 0
for ch in text:
    if ch.isupper():
        count += 1
print("Uppercase letters =", count)
```

```python
'''Question 10: Write a program to count the number of lowercase letters in a string.'''
text = input("Enter a string ")
count = 0
for ch in text:
    if ch.islower():
        count += 1
print("Lowercase letters =", count)
```

```python
'''Question 11: Write a program to display all elements of a list.'''
numbers = [10, 20, 30, 40, 50]
for num in numbers:
    print(num)
```

```python
'''Question 12: Write a program to find the minimum element in a list.'''
numbers = [25, 10, 45, 5, 30]
smallest = min(numbers)
print("Minimum Element =", smallest)
```
