```python
str1 = input("enter the string")
alpha = 0
digit = 0
special= 0
for i in str1:
    if("A"<=i<="Z")or("a"<=i<="z"):
        alpha+=1
    elif("0"<=i<="9"):
        digit+=1
    else:
        special+=1
print("alphabet=",alpha)
print("digit=",digit)
print("special=",special)
        
        
```

```python
str1 = input("Enter a string: ")

result = ""

for i in str1:
    if i == 'a':
        result = result + 'A'
    elif i == 'b':
        result = result + 'B'
    elif i == 'c':
        result = result + 'C'
    elif i == 'd':
        result = result + 'D'
    elif i == 'e':
        result = result + 'E'
    elif i == 'f':
        result = result + 'F'
    elif i == 'g':
        result = result + 'G'
    elif i == 'h':
        result = result + 'H'
    elif i == 'i':
        result = result + 'I'
    elif i == 'j':
        result = result + 'J'
    elif i == 'k':
        result = result + 'K'
    elif i == 'l':
        result = result + 'L'
    elif i == 'm':
        result = result + 'M'
    elif i == 'n':
        result = result + 'N'
    elif i == 'o':
        result = result + 'O'
    elif i == 'p':
        result = result + 'P'
    elif i == 'q':
        result = result + 'Q'
    elif i == 'r':
        result = result + 'R'
    elif i == 's':
        result = result + 'S'
    elif i == 't':
        result = result + 'T'
    elif i == 'u':
        result = result + 'U'
    elif i == 'v':
        result = result + 'V'
    elif i == 'w':
        result = result + 'W'
    elif i == 'x':
        result = result + 'X'
    elif i == 'y':
        result = result + 'Y'
    elif i == 'z':
        result = result + 'Z'
    else:
        result = result + i

print("Uppercase String =", result)
```
