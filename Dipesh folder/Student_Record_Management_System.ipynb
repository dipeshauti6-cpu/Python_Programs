```python
Roll_no = input("Enter Roll Number to Search =")
file = open("student.txt", "r")
found = False
for line in file:
    data = line.strip().split(",")
    roll = data[1].split("=")[1].strip()
    if roll == Roll_no:
        print("Student Found")
        print(data[0])
        print(data[1])
        print(data[2])
        found = True
        break
if found == False:
    print("Student Not Found")
file.close()
```

```python
Roll_no = input("Enter Roll Number =")
new_marks = input("Enter New Marks =")

file = open("student.txt", "r")
lines = file.readlines()
file.close()

file = open("student.txt", "w")

for line in lines:
    data = line.strip().split(",")

    old_roll = data[1].split("=")[1].strip()

    if old_roll == Roll_no:
        file.write(data[0] + "," + data[1] + ", Mark = " + new_marks + "\n")
    else:
        file.write(line)

file.close()

print("Marks Updated Successfully.")
```

```python
roll = input("Enter Roll Number to Delete =")
file = open("student.txt", "r")
lines = file.readlines()
file.close()

file = open("student.txt", "w")

found = False

for line in lines:
    data = line.strip().split(",")
    student_roll = data[1].split("=")[1].strip()
    if student_roll != roll:
        file.write(line)
    else:
        found = True

file.close()

if found:
    print("Record Deleted Successfully")
else:
    print("Record Not Found")
```

```python
file = open("student.txt", "r")
lines = file.readlines()
file.close()
for i in range(len(lines)):
    for j in range(i + 1, len(lines)):
        mark1 = int(lines[i].split(",")[2].split("=")[1])
        mark2 = int(lines[j].split(",")[2].split("=")[1])
        if mark1 < mark2:
            temp = lines[i]
            lines[i] = lines[j]
            lines[j] = temp
print("Student Records Sorted By Marks")
for line in lines:
    print(line)
```

```python
file = open("student.txt", "r")
lines = file.readlines()
file.close()
lines.sort(key=lambda x: int(x.split(",")[2].split("=")[1]), reverse=True)
file = open("top3.txt", "w")
for i in range(3):
    file.write(lines[i])
file.close()
print("Top 3 Students Exported Successfully.")
```

```python
file = open("student.txt", "r")
count1 = 0
count2 = 0
for line in file:
    data = line.strip().split(",")
    marks = int(data[2].split("=")[1])
    if marks > 90:
        count1 = count1 + 1
    if marks < 35:
        count2 = count2 + 1
file.close()
print("Students Scored More Than 90 =", count1)
print("Students Scored Less Than 35 =", count2)
```

```python
while True:
    print("1.Add\n2.View\n3.Search\n4.Update\n5.Delete\n6.Exit")
    ch = int(input("Enter your choice: "))
    if ch == 1:
        name = input("Enter Name: ")
        roll = input("Enter Roll No: ")
        marks = input("Enter Marks: ")
        f = open("Student_MenuDriven.txt", "a")
        f.write("Name: " + name + ", Roll No: " + roll + ", Marks: " + marks + "\n")
        f.close()
    elif ch == 2:
        f = open("Student_MenuDriven.txt", "r")
        print(f.read())
        f.close()
    elif ch == 3:
        roll = input("Enter Roll No: ")

        f = open("Student_MenuDriven.txt", "r")
        for line in f:
            if "Roll No: " + roll in line:
                print(line)
        f.close()
    elif ch == 4:
        roll = input("Enter Roll No: ")
        old = input("Enter Old Marks: ")
        new = input("Enter New Marks: ")
        f = open("Student_MenuDriven.txt", "r")
        data = f.read()
        f.close()
        data = data.replace("Roll No: " + roll + ", Marks: " + old,
                            "Roll No: " + roll + ", Marks: " + new)
        f = open("Student_MenuDriven.txt", "w")
        f.write(data)
        f.close()
    elif ch == 5:
        roll = input("Enter Roll No: ")
        f = open("Student_MenuDriven.txt", "r")
        lines = f.readlines()
        f.close()
        f = open("Student_MenuDriven.txt", "w")
        for line in lines:
            if "Roll No: " + roll not in line:
                f.write(line)
        f.close()
    elif ch == 6:
        print("Program Ended")
        break
    else:
        print("Invalid Choice")
```
