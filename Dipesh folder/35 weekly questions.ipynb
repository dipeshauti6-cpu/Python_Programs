```python
#1)POSITIONAL ARUGMENT:
def subtract(a,b):
    print("Subtraction=",a-b)
subtract(10,5)
```

```python
def student(name,age):
    print("Name=",name)
    print("Age=",age)
student("Dipesh",18)
```

```python
def multiply(a,b,c):
    print("Multiplication=",a*b*c)
multiply(5,5,5)
```

```python
def name(first,last):
    print("Fullname=",first,last)
name("Dipesh","Auti")
```

```python
def area(length,width):
    print("Area of Rectangle=",length*width)
area(54,64)
```

```python
#2)KEYWORD ARUGMENT:
def employee(name,salary,city):
    print("Name=",name)
    print("Salary=",salary)
    print("City=",city)
employee(name="Dipesh",salary=300000,city="New York")
```

```python
def add(a,b,c):
    print("Addition=",a+b+c)
add(b=20,a=10,c=20)
```

```python
def product_detail(id,name,price):
    print("ID=",id)
    print("Name=",name)
    print("Price=",price)
product_detail(name="XYZ",price=500,id=121)
```

```python
def interest(p,r,t):
    print("Simple_Interest=",p*r*t/100)
interest(p=50000,r=10,t=2)
```

```python
def user(name,age,city):
    print("Name=",name)
    print("Age=",age)
    print("City=",city)
user(name="Dipesh",age=18,city="New York")
```

```python
#3)Default Parameters:
def add(a,b=5):
    print("Addition=",a+b)
add(10)
```

```python
def Power(a,b=2):
    print("Power=",a**b)
Power(5)
```

```python
def greet_user(name="Guest"):
    print("Hello",name)
greet_user()
```

```python
def salary(normal,bonus=1000):
    print("Total_Salary=",normal+bonus)
salary(30000)
```

```python
def circle(radius=1):
    print("Area of Circle=",3.14*radius*radius)
circle()
```

```python
#4)Variable Length Positional Arguments (*args):
def add(*num):
    print("Addition=",sum(num))
add(10,20,30)
```

```python
def maximum(*num):
    print("Maximum=",max(num))
maximum(10,20,30,40)
```

```python
def marks(*mark):
    print("Marks of Student=",mark)
marks(80,90,80,90)
```

```python
def avg(*num):
    print("Average=",sum(num)/len(num))
avg(10,20,30,40)
```

```python
def count(*arg):
    print("Total_Argument=",len(arg))
count(1,2,3,4,5,6)
```

```python
#5) Variable Length Keyword Arguments (**kwargs):
def student(**data):
    print(data)
student(Name="Dipesh",Age=18)
    
```

```python
def details(**data):
    for k,v in data.items():
        print(k,"=",v)
details(name="Dipesh",age=18)
```

```python
def total(**item):
    print("Total=",sum(item.values()))
total(book=100,pen=10,bag=1000)
```

```python
def user(**profile):
    print(profile)
user(Name="Dipesh",Age=18,City="New York")
```

```python
def user(**profile):
    for k,v in profile.items():
        print(k,"=",v)
user(Name="Dipesh",Age=18,City="New York")
```

```python
#6) Positional Only Arguments:
def add(a,b,/):
    print("Addition=",a+b)
add(10,20)
```

```python
def divide(a,b,/):
    print("Division=",a/b)
divide(10,5)
```

```python
def display(a,b,/):
    print(a,b)
display("Hello","Bacho")
```

```python
def sqaure(side,/):
    print("Area of Sqaure=",side*side)
sqaure(6)
```

```python
def values(a,b,/):
    print("Value=",a)
    print("Value=",b)
values(100,200)
```

```python
#7) Keyword Only Arguments:
def add(*,a,b):
    print("Addition=",a+b)
add(a=10,b=10)
```

```python
def emp(*,name,age,city):
    print("Name=",name)
    print("age=",age)
    print("City=",city)
emp(name="Dipesh",age=18,city="New York")
```

```python
def total_marks(*,m1,m2,m3):
    print("Total_Marks=",m1+m2+m3)
total_marks(m1=80,m2=85,m3=90)
```

```python
def detail(name,*,age):
    print("Name=",name)
    print("Age=",age)
detail("Dipesh",age=18)
```

```python
def discount(price,*,discount):
    print("Total_Price=",price-discount)
discount(2000,discount=1000)
```
