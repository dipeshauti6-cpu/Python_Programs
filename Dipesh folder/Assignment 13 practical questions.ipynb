```python
import string
text = input("enter the text=")
letters = 0 
digits = 0
spaces = 0
for i in text:
    if i in string.ascii_letters:
        letters+=1
    elif i in string.digits:
        digits+=1
    elif i==" ":
        spaces+=1
    else:
        break
print("Letters =",letters)
print("Digits=",digits)
print("Spaces=",spaces)
```
