```python
''' Question22: Write a python program using while loop to print number from 1 to 10.'''
i = 1
while i<=10:
    print(i)
    i+=1
```

```python
''' Question23: Write a python program using while loop to print even number from 2 to 20.'''
i = 2
while i<=20:
    print(i)
    i+=2
```

```python
''' Question24: Write a python program using while loop to calculate the sum of number from 1 to 10.'''
i = 1
total=0
while i<=10:
    total+=i
    i+=1
print("Sum=",total)
```

```python
''' Question25: Write a python program using while loop to print multiplication table of 5.'''
i = 1
while i <= 10:
    print("5 x", i, "=", 5 * i)
    i += 1
```

```python
''' Question28: Write a nested loop program to print: *, **, ***, ****.'''
for i in range(1,5):
    for j in range(i):
        print("*",end="")
    print()
```

```python
''' Question29: Write a nested loop program to print: 1, 12, 123, 1234.'''
for i in range (1,5):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
''' Question30: Write a program using a while loop to find the factorial of a number entered by the user.'''
num = int(input("enter a number"))
fact = 1
i = 1
while i<=num:
    fact= fact*i
    i+=1
print("factorial=",fact)
```
