```python

while True:
    print("\n1.Enter String \n2. Convert to Uppercase \n3. Convert to Lowercase \n4. Display \n5. Exit")
    choice = int(input("Enter your choice= "))
    if choice == 1:
        string = input("Enter a string= ")
        print("String inserted successfully")
    elif choice == 2:
        print(string.upper())
        print("Converted successfully")
    elif choice == 3:
        print(string.lower())
        print("Converted successfully")
    elif choice == 4:
        print("String =", string)
        print("Displayed successfully")
    elif choice == 5:
        print("Thanks for visiting")
        break
    else:
        print("Enter valid operation")
```

```python
while True:
    print("\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Division\n5.Modulus\n6.Exit")
    choice = int(input("Enter your choice="))
    if choice == 1:
        num1 = float(input("Enter first number="))
        num2 = float(input("Enter second number="))
        print("Answer =", num1 + num2)
    elif choice == 2:
        num1 = float(input("Enter first number="))
        num2 = float(input("Enter second number="))
        print("Answer =", num1 - num2)
    elif choice == 3:
        num1 = float(input("Enter first number= "))
        num2 = float(input("Enter second number= "))
        print("Answer =", num1 * num2)
    elif choice == 4:
        num1 = float(input("Enter first number= "))
        num2 = float(input("Enter second number= "))
        if num2 != 0:
            print("Answer =", num1 / num2)
        else:
            print("Division by zero is not possible")
    elif choice == 5:
        num1 = float(input("Enter first number="))
        num2 = float(input("Enter second number= "))
        print("Answer =", num1 % num2)
    elif choice == 6:
        print("Thanks for visiting")
        break
    else:
        print("Enter valid operation")
```
