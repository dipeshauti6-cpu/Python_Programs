```python
'''Question 28: Write a program to check if a year is a leap year or not.'''
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
'''Question 30: Write a program to check if a person is eligible for a bank loan using nested if-else: Age must be between 21 and 60.
Monthly salary must be greater than 25,000. Handle each condition separately and print a suitable message.'''
age = int(input("Enter age: "))
salary = int(input("Enter monthly salary: "))
if age >= 21 and age <= 60:
    if salary > 25000:
        print("Eligible for Loan")
    else:
        print("Salary is too low")
else:
    print("Age criteria not satisfied")
```

```python
'''Question 31: Write a program that takes two numbers and an operator (+, -, *, /) as input and performs the operation using if-elif-else.
Note: Handle division by zero using a nested if inside the division condition.'''
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")
if op == "+":
    print("Result =", num1 + num2)
elif op == "-":
    print("Result =", num1 - num2)
elif op == "*":
    print("Result =", num1 * num2)
elif op == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid Operator")
```

```python
'''Question 32: Write a program to categorize a person's BMI using if-elif-else.
Below 18.5 → Underweight
18.5 – 24.9 → Normal
25.0 – 29.9 → Overweight
30 and above → Obese
BMI = weight / (height * height)'''
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meter: "))
bmi = weight / (height * height)
if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

```python
'''Question 33: Write a program to check if three sides form a valid triangle.
If valid, classify it using nested if-else:
All sides equal → Equilateral
Any two sides equal → Isosceles
All sides different → Scalene'''
a = int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == b and b == c:
        print("Equilateral Triangle")
    elif a == b or b == c or a == c:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")
else:
    print("Invalid Triangle")
```

```python
'''Question 34: Write a program for a simple ATM system using if-elif-else and nested if-else.
If entered PIN is correct (PIN = 1234), ask for withdrawal amount.
If amount is less than or equal to balance (balance = 10000), deduct it and show remaining balance.
If amount exceeds balance, print 'Insufficient Balance'.
If PIN is wrong, print 'Invalid PIN'.'''
pin = int(input("Enter PIN: "))
balance = 10000
if pin == 1234:
    amount = int(input("Enter withdrawal amount: "))
    if amount <= balance:
        balance = balance - amount
        print("Withdrawal Successful")
        print("Remaining Balance =", balance)
    else:
        print("Insufficient Balance")
else:
    print("Invalid PIN")
```

```python
'''Question 35: Write a program for an Electricity Bill Calculator using if-elif-else.
0 – 100 units → Rs. 1.50 per unit
101 – 300 units → Rs. 2.00 per unit
301 – 500 units → Rs. 3.00 per unit
Above 500 units → Rs. 5.00 per unit
Add a fixed charge of Rs. 50 and print the total bill amount.'''
units = int(input("Enter units consumed: "))
if units <= 100:
    bill = units * 1.50
elif units <= 300:
    bill = units * 2.00
elif units <= 500:
    bill = units * 3.00
else:
    bill = units * 5.00
total_bill = bill + 50
print("Total Bill Amount =", total_bill)
```
