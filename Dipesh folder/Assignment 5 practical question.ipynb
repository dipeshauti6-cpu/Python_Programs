```python
'''Question 24:Write a program to print numbers from 1 to 10 using a for loop.'''
for i in range(1, 11):
    print(i)
```

```python
'''Question 25:Write a program to print numbers from 10 to 1 using a while loop.'''
i = 10
while i >= 1:
    print(i)
    i -= 1
```

```python
'''Question 26:Write a program to print all even numbers from 1 to 50.'''
for i in range(2,50):
    if i%2==0:
        print(i)
```

```python
'''Question 27:Write a program to find the sum of numbers from 1 to n.'''
n = int(input("Enter a number"))
sum = 0
for i in range(1, n + 1):
    sum += i
print("Sum =", sum)
```

```python
'''Question 28:Write a program to print the multiplication table of a given number.'''
num = int(input("Enter a number"))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
'''Question 29:Write a program to count the number of digits in a given number.'''
num = int(input("Enter a number"))
count = 0
while num > 0:
    count += 1
    num //= 10
print("Number of digits =", count)
```

```python
'''Question 30:Write a program to reverse a given number using a loop.'''
num = int(input("Enter a number"))
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10
print("Reverse number =", reverse)
```

```python
'''Question 31:Write a program to find the factorial of a number using a loop.'''
num = int(input("Enter a number"))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)
```

```python
''' Question 32:Write a program that stops the loops when the value reaches 5 using break.'''
for i in range (1,11):
    if i == 5:
        break
    print(i)
```

```python
'''Question 33:Write a program that skips the number 5 in a loop using continue.'''
for i in range (1,11):
    if i==5 :
        continue
    print(i)    
```

```python
'''Question 34:Write a program that uses pass inside a loop as a placeholder.'''
for i in range(1, 6):
    pass
print("sum=", 10*6)
```

```python
'''Question 35:Write a program that keeps asking for user input until the user enters 0.'''
num = 5
while num != 0:
    num = int(input("Enter a number (0 to stop)"))
print("Program Ended")
```

```python
'''Question 36:Write a program to check whether a given number is prime.'''
num = int(input("Enter a number"))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```
