```python
'''Question 23: Write a program to count how many even and odd numbers are present from 1 to n.'''
n = int(input("Enter a number: "))
even = 0
odd = 0
for i in range(1, n + 1):
 if i % 2 == 0:
   even += 1
 else:
   odd += 1
print("Even numbers =", even)
print("Odd numbers =", odd)
```
